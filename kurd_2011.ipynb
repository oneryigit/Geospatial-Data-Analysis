```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

#sns.set_theme(style='white')
```

```python
df=pd.read_stata('kurd_r_ucpd.dta')
```

```python
df_d=df[['zor_gocguv2', 'event_n', 'border_city', 'education', 'soru39','soru41', 'soru45','dindarlik']]

df_d
```

```python
map1= {"Erkek":1, "Kadin":0}

df_d['soru39']=df_d['soru39'].map(map1)
df_d['soru39']=df_d['soru39'].astype('int')

df_d.head()
```

```python
df_f=df_d[['zor_gocguv2', 'event_n', 'border_city', 'education', 'soru39','soru41', 'soru45','dindarlik']]

df_f
```

```python
#df_f[['soru24','soru31']]=df_f[['soru24','soru31']].apply(lambda  x: pd.factorize(x)[0])
```

```python
corr=df_f.corr()
```

```python
corr
```

```python
# Increase the size of the heatmap.
mask = np.triu(np.ones_like(corr, dtype=bool))
cmap=sns.color_palette("vlag", as_cmap=True)
xlabels=['Forced Displacement','Number of Conflict', 'Border City', 'Education','Gender','Age','Income','Religiosity']

plt.style.use('seaborn-colorblind')

plt.rcParams.update({'font.size': 14}) #to set font size for all properties of plot
plt.rcParams.update({'font.sans-serif':'Times New Roman'}) #or 'helvetica'

fig,ax= plt.subplots(figsize=(8, 10))


# Store heatmap object in a variable to easily access it when you want to include more features (such as title).
# Set the range of values to be displayed on the colormap from -1 to 1, and set the annotation to True to display the correlation values on the heatmap.
heatmap = sns.heatmap(corr, vmin=-1, vmax=1, annot=True,  fmt=".2f",mask=mask, cmap=cmap,center=0,
                      square=True, linewidths=.5, cbar_kws={"shrink": .6},xticklabels=xlabels, yticklabels=xlabels)

# Give a title to the heatmap. Pad defines the distance of the title from the top of the heatmap.
#heatmap.set_title('Correlation Matrix', fontdict={'fontsize':16}, pad=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

ax.grid(False)

#plt.axis('off')
fig.tight_layout()
ax.set_facecolor("w")
fig.set_facecolor('w')
plt.show()
fig.savefig('corr_plot.png',bbox_inches='tight', dpi=300);
fig.savefig('corr_plot.pdf',bbox_inches='tight', dpi=300);
```
